```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
kids = {'Height': [65.5, 71.52, 69.40, 71.22, 67.79, 71.70, 69.80, 70.01, 67.90, 66.78, 66.49, 67.62, 68.30, 67.12, 68.28],
        'Weight': [112.99, 136.49, 123.03, 132.34, 124.30, 148.30, 141.49, 136.46, 122.37, 120.67, 127.45, 124.14, 125.61, 122.46, 116.09]}

dfkids=pd.DataFrame(kids)
print(dfkids)
dfkids.head()
```

```python
dfkids.info()
```

```python
dfkids.describe()
```

```python
#Scatterplot of Height vs Weight
plt.scatter(dfkids['Height'], dfkids['Weight'], marker='*')
plt.xlabel('Height in Inches')
plt.ylabel('Weight in Pounds')
plt.title('Height vs Weight')
plt.show()
```

```python
# Here i tried using height in meters and weight in Kilograms as those are the units i am most familiar with
# 1 inch = 0.0254 meters
# 1 pound = 0.453592 kilograms

# Convert height from inches to meters
dfkids['Height_m'] = dfkids['Height'] * 0.0254

# Convert weight from pounds to kilograms
dfkids['Weight_kg'] = dfkids['Weight'] * 0.453592

# Scatterplot of Height vs Weight in meters and kilograms
plt.scatter(dfkids['Height_m'], dfkids['Weight_kg'])
plt.xlabel('Height in Meters')
plt.ylabel('Weight in Kilograms')
plt.title('Height vs Weight')
plt.show()
```

```python
X = dfkids[['Height']]
y = dfkids['Weight']

model = linear_model.LinearRegression()
model.fit(X, y)

model.coef_
model.intercept_
# Predict weight for a given height
new_height = [[70]]
predicted_weight = model.predict(new_height)
print("Predicted weight for height 170:", predicted_weight)
```

```python
# i also did prediction and also changed the height to meters and weight to kgs

X = dfkids[['Height_m']]  # Using height in meters
y = dfkids['Weight_kg']   # Using weight in kilograms

model = linear_model.LinearRegression()
model.fit(X, y)

print("Coefficient (slope):", model.coef_)
print("Intercept:", model.intercept_)

# Predict weight for a given height
new_height_m = [[1.71]]  # Height in meters
predicted_weight_kg = model.predict(new_height_m)
print("Predicted weight for height 1.78 meters:", predicted_weight_kg)
```
